```python
import pandas as pd #Manipulasi dan analisis data
from matplotlib import pyplot as plt

data = pd.read_csv("vgsales.csv")

data
```

```python
data['Platform'].value_counts() #Menampilkan Nilai Data Platform
```

## Diagram Batang

```python
x = list(data.iloc[:,6])
y = list(data.iloc[:,10])

plt.bar(x, y, color='red') #Jumlah Sales NA dan Sales Global
plt.title("Jumlah Sales")
plt.xlabel("NA_Sales")
plt.ylabel("Global_Sales")
```

## Diagram Garis atau Grafik Plot dan garis

```python
data.head()
plt.plot(data.Year, data.Global_Sales)
plt.title("Penjualan tahunan")
plt.xlabel("Tahun")
plt.ylabel("Penjualan Global")

#Visual Data Global Sales Setiap Tahun
```

## Diagram Pie


```python
labels = data['Genre'].value_counts().rename_axis('Game').reset_index(name='Penjualan')
labels = pd.DataFrame(labels)
labels.head(10)
```

```python
plt.figure(figsize=(10,10))
plt.pie(data['Genre'].value_counts(), labels=labels['Game'], autopct='%.1f%%', startangle=90)

plt.title('Penjualan Global')
plt.show() #PIE Chart GLobal Sales

```
